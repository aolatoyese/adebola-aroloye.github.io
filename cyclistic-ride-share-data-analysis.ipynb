# Process and Analyze Data

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# Import python libaries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt 
import datetime 

# Input data files are available in the read-only "../input/" directory
```

```python
# Import 12 CSV file 

Jan_2022 = pd.read_csv('../input/cyclistic-data/202201-divvy-tripdata.csv')
Feb_2022 = pd.read_csv('../input/cyclistic-data/202202-divvy-tripdata.csv')
Mar_2022 = pd.read_csv('../input/cyclistic-data/202203-divvy-tripdata.csv')
Apr_2022 = pd.read_csv('../input/cyclistic-data/202204-divvy-tripdata.csv')
May_2022 = pd.read_csv('../input/cyclistic-data/202205-divvy-tripdata.csv')
Jun_2022 = pd.read_csv('../input/cyclistic-data/202206-divvy-tripdata.csv')
Jul_2022 = pd.read_csv('../input/cyclistic-data/202207-divvy-tripdata.csv')
Aug_2022 = pd.read_csv('../input/cyclistic-data/202208-divvy-tripdata.csv')
Sep_2022 = pd.read_csv('../input/cyclistic-data/202209-divvy-publictripdata.csv')
Oct_2022 = pd.read_csv('../input/cyclistic-data/202210-divvy-tripdata.csv')
Nov_2022 = pd.read_csv('../input/cyclistic-data/202211-divvy-tripdata.csv')
Dec_2022 = pd.read_csv('../input/cyclistic-data/202212-divvy-tripdata.csv')
```

```python
# Merge and stack all 12 months using concat

df= pd.concat([Jan_2022,Feb_2022,Mar_2022,Apr_2022,May_2022,Jun_2022,Jul_2022,Aug_2022,Sep_2022,Oct_2022,Nov_2022,Dec_2022], ignore_index = True)
```

```python
#dataframe

df 
```

```python
# Get info about dataframe 

df.info()
```

```python
# Change data type in dataframe 

df["started_at"]= pd.to_datetime(df["started_at"])
df["ended_at"]= pd.to_datetime(df["ended_at"])
df.info()
```

```python
# Calculate a new column as ride_lenght

df["ride_length"]= (df["ended_at"] - df["started_at"]).dt.total_seconds().div(60).astype(float)
df["ride_length"]
```

```python
# Remove column not needed for this analysis

df.drop(['start_station_name','start_station_id','end_station_name', 'end_station_id' ,'start_lng', 'start_lat', 'end_lat' , 'end_lng'] , axis = 1 , inplace = True)
```

```python
# Revelant columns for analysis
df
```

```python
# adding new column as day_of_week 
# day_of_week

df['day_of_week'] = df['started_at'].dt.day_name()
df
```

```python
# adding new column as  month

df['Month'] = df['started_at'].dt.month_name()
df
```

```python
# cleaning columns with negative ride_length
# info
print(len(df[df['ride_length']<0]))
print(len(df[df['ride_length']>0]))
```

```python
# cleaning columns with negative ride_length

df.drop( df[df['ride_length']<0].index , inplace = True)
df
```

```python
# Negative ride_length cleaned out

print(len(df[df['ride_length']<0]))
print(len(df[df['ride_length']>0]))
```

```python
#Plot the number of rides for each type casual or membership
df["member_casual"].value_counts().plot(kind='bar', figsize=(6, 5), rot=0, color=['darkorange', 'steelblue'])
plt.xlabel("Type of User", labelpad=14)
plt.ylabel("Count of Rides", labelpad=14)
plt.title("Numbers of Customers", y=1.02);
```

```python
#Plot the rides by day with how many casuals memebers and membership per day 
df.groupby(["day_of_week", "member_casual"]).size().unstack().plot(kind="bar", rot=0)
plt.xlabel("Days of The Week", labelpad=14)
plt.ylabel("Count of Rides", labelpad=14)
plt.title("Count of Rides by Day", y=1.02)
fig= plt.gcf()
fig.set_size_inches(12, 5.5) ;
```

```python
#Plot the rides by day with how many casuals memebers and membership per day 
df.groupby(["Month", "member_casual"]).size().unstack().plot(kind="line", rot=0)
plt.xlabel("Months", labelpad=14)
plt.ylabel("Number of Rides", labelpad=14)
plt.title("Count of Rides by month", y=1.02)
fig= plt.gcf()
fig.set_size_inches(12, 6.5) ;
```

```python
#Plot of the type of bike used more by members and casual 
df.groupby(["rideable_type", "member_casual"]).size().unstack().plot(kind="bar", rot=0)
plt.xlabel("Types of Bikes", labelpad=14)
plt.ylabel("Count of Rides per Type", labelpad=14)
plt.title("Count of Rides by Type of Bike", y=1.02)
fig= plt.gcf()
fig.set_size_inches(10, 5.5) ;
```
